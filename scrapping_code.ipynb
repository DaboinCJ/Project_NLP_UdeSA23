These are the links of the presidential debates videos we're going to extract scripts from...

````python
## We make sure tjhe youtube transcripts library is available
from youtube_transcript_api import YouTubeTranscriptApi

## We create a function to cut the videos by mm:ss
def time_to_seconds(time_str):
    """Convertir un string de tiempo (formato 'h.mm:ss' o 'mm:ss') a segundos."""
    parts = list(map(int, time_str.split(':')))

    if len(parts) == 3:
        h, m, s = parts
        return h * 3600 + m * 60 + s
    elif len(parts) == 2:
        m, s = parts
        return m * 60 + s
    else:
        raise ValueError("Formato de tiempo no reconocido.")
````
